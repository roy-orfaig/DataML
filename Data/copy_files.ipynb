```python
import os
import shutil

# Copy all files from source to target
def copy_files(source, target):
    for item in os.listdir(source):
        source_path = os.path.join(source, item)
        target_path = os.path.join(target, item)
        
        if os.path.isfile(source_path):  # Copy only files
            shutil.copy2(source_path, target_path)
            print(f'Copied: {source_path} -> {target_path}')


```

```python
source='/isilon/Automotive/RnD/roy.o/broken_part_dataset/2025-03-07_11-01-53/images/train'
target='/isilon/Automotive/RnD/roy.o/workspace/data/datasets/borken_part/2025-02-18_08-38-20/images/train'
copy_files(source, target)

print("File copying completed!")
```
